```python
import numpy as np
import matplotlib.pyplot as plt

import os
import sys
module_path = os.path.abspath(os.path.join('..')) # need to go up one level in project
if module_path not in sys.path:
    sys.path.append(module_path)
#This allows you to import the desired function from the module hierarchy:

import pipeline.PreProcessing as pp
```

```python

path = "../data/testData_1/"
X3D, Y = pp.load_datasets(path, 0)

print("X3D.shape"+str(X3D.shape))

#extract n*n images
ni = 3
nj = 3
sti = 1 #start from sti-th line if too many images

%matplotlib inline


if len(X3D)>=ni*nj:
    fig, axs = plt.subplots(ni-sti, nj)
    for i in range(sti,ni):#or range(ni-100:ni)
        for j in range(0,nj):
            axs[i-sti,j].imshow((X3D[nj*i+j].sum(axis=2)))    
    #plt.tight_layout()
else:
    print("too many plots requested")
```
